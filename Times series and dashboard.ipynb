```python
df.set_index( "Time_period" , inplace= True )
```

```python
from prophet import Prophet
```

```python
def create_fp_df(burden):
  fp_df = df[[burden]].copy()
  fp_df.reset_index(inplace = True)
  fp_df.rename(columns = {'Time_period':'ds'}, inplace = True)
  fp_df.rename(columns = {burden:'y'}, inplace = True)
  return fp_df
```

```python
fp_df = create_fp_df('Stunting')
```

```python
# Initialize the Model
model=Prophet()
```

```python
model.fit(fp_df)
```

```python
# Predicting next 5 years
future_dates=model.make_future_dataframe(5, freq='A')
```

```python
future_dates.tail()
```

```python
prediction=model.predict(future_dates)
```

```python
### plot the predicted projection
model.plot(prediction,include_legend=True)
plt.title("Stunting")
```

# Dashboard

```python
df1= pd.read_csv("timeseries_df.csv")
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import pandas as pd
import plotly.express as px
```

```python
# Initialize the Dash app
app = dash.Dash(__name__)
```

```python
# Define the layout of the dashboard
app.layout = html.Div([
    html.H1("Time Series Dashboard"),
    
    # Dropdown to select the column to visualize
    dcc.Dropdown(
        id='column-selector',
        options=[{'label': col, 'value': col} for col in df1.columns[1:]],  # Exclude 'Year'
        value='Stunting',  # Default selection
        multi=False
    ),
    
    # Line chart to display the selected column's data
    dcc.Graph(id='line-chart'),
])
```

```python
# Define callback to update the line chart based on user input
@app.callback(
    Output('line-chart', 'figure'),
    [Input('column-selector', 'value')]
)
def update_line_chart(selected_column):
    fig = px.line(df1, x='Time_period', y=selected_column, title=f'Time Series for {selected_column}')
    return fig
```

```python
# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
